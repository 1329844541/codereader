# 美剧对婴儿姓名的影响

```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib as mpl
mpl.style.use('ggplot')
```

```python
data = pd.read_csv('./dataset/NationalNames.csv')
```

```python
def plotname(name,sex):
    data_named=data[(data.Name==name) & (data.Gender==sex)]
    plt.figure(figsize=(12,8))
    plt.plot(data_named.Year,data_named.Count,'g-')
    plt.title('%s name variation over time'%name)
    plt.ylabel('counts')
    plt.xticks(data_named.Year,rotation='vertical')
    plt.xlim([1985,2015])    
    plt.show() 
```

```python
characters = [('Rachel','F'),('Monica','M'),('Phoebe','F'),('Ross','F'),('Chandler','F'),('Joey','F'), ('Khaleesi', 'F')]
```

```python
for character in characters:
    plotname(character[0],character[1])
```
